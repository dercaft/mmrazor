```python
import os
import csv
import json
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATH='/data/work_dirs/wyh/hybrid/change_each_channel_1tofull'
SUFFIX='230509_160143'
filepath="_".join([PATH,SUFFIX])
filename="total.csv"
with open(os.path.join(filepath,filename)) as f:
    f_csv=csv.reader(f)
    headers=next(f_csv)
    data=[row for row in f_csv]
group_name=[row[0] for row in data]
channel=[int(row[1]) for row in data]
data=[np.array(row[2:],dtype=np.float32) for row in data]
# data=np.array(data,dtype=np.float32)
x=[ [j/channel[i] for j,v in enumerate(row) ]for i, row in enumerate(data) ]
```

```python
for i,row,x in zip(range(len(group_name)),data,x):
    plt.plot(x,row,label=group_name[i])
plt.xlabel("channel ratio")
plt.ylabel("CKA")
plt.title("Resnet20's CKA changed with each channel")
plt.legend()
plt.show()
```
